```python
from pycaret.classification import *
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\modeling\onlyonehot\onlyonehot.csv",sep=";")
data.drop(columns="Unnamed: 0",inplace=True)
data
```

```python
# tag 데이터 불러오기 
tag_path = r"C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\merge_dataset\tag_merge.csv"
tag_df = pd.read_csv(tag_path)
tag_df
```

```python
# tag 데이터 binary 변환 
tag_df.iloc[:,1:] = tag_df.iloc[:,1:].astype("bool")
tag_df.iloc[:,1:] = tag_df.iloc[:,1:].astype("int")
```

```python
data = pd.merge(data,tag_df,how="inner",left_on="appid",right_on="appid")
```

```python
# tag 개수 컬럼 추가 
data["tag_num"] = tag_df.iloc[:,1:].sum(axis=1)
```

```python
# 모델 학습에 불필요한 컬럼 제거 
data.drop(columns=["genre","appid"],inplace=True)
```

```python
# 정답지 분리 
X = data.drop(columns="Review")
y = data["Review"]
```

```python
# train, validation 분리 
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state = 1234, 
    stratify = y 
)
```

```python
# train 데이터 조정 : pycaret에서 target으로 y_train 지정을 위함 
X_train["Review"] = y_train.to_list()
```

```python
# pycaret classification 객체 생성 
exp = ClassificationExperiment()
```

```python
# setup : stratified kfold, target 불균형 해결을 위한 SMOTE 사용(fix_imbalance) 
exp.setup(X_train, target='Review', session_id=123, fold_strategy="stratifiedkfold",fix_imbalance=True)
```

```python
# 5개 모델 불러오기 
ada_model = load_model(r'C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\finaldata\ada_tune_model')
rf_model = load_model(r'C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\finaldata\rf_tune_model')
lgbm_model = load_model(r'C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\finaldata\lgbm_tune_model')
et_model = load_model(r'C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\finaldata\et_tune_model')
gbc_model = load_model(r'C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML\web_crawling\CGW\finaldata\gbc_tune_model')
```

```python
# soft voting 
tune_blender = exp.blend_models([rf_model,lgbm_model,gbc_model,et_model,ada_model],method="soft")
```

```python
# prediction
prediction = exp.predict_model(tune_blender, data = X_test)
prediction
```

```python
b = prediction["prediction_label"]
b
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,b))
```
