```python
import pandas as pd
import numpy as np
```

```python
raw = pd.read_csv(r'C:\Users\NT550-045\Desktop\ml2\2022_steam2.csv', sep='\t')
raw.head(1)
```

```python
raw.info()
```

```python
raw['tags'] = raw['tags'].apply(lambda x: eval(x) if x != '[]' else set())
raw['tags']
```

```python
tag_set = set()
for i in range(len(raw)):
    tag_set |= set(raw.loc[i,'tags'])
```

```python
len(tag_set)
```

```python
from collections import deque
tag_dict = {tag : deque([]) for tag in tag_set}
```

```python
for i in range(len(raw)):
    for tag in tag_set:
        try:
            tag_dict[tag].append(raw.loc[i,'tags'][tag])
        except:
            tag_dict[tag].append(0)
```

```python
tag_dict['appid'] = raw['appid']
```

```python
tag_df = pd.DataFrame(tag_dict)
```

```python
col = ['appid'] + tag_df.columns[:-1].to_list()
tag_df = tag_df[col]
```
