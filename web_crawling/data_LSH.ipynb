```python
from collections import deque
import numpy as np
```

```python
z = ["Developer","ValidOSList","community_visible_stats","Publisher",
                "IsFreeApp","Exclude from family sharing","has_adult_content",
                "Store Release Date","Primary Genre","Store Genres","Is free on the store?",
                "market_presence","review_score","review_percentage","SteamChinaApproved",
                "has_adult_content_violence","ReleaseStateSteamChina"]

y = {col : deque([]) for col in z}
y['appid'] = appid
```

```python
y
```

```python
import time
```

```python
for idx, app in enumerate(appid, start=1):
    if idx % 50 == 0:
        time.sleep(60)
    print(idx, app)
    res = s.get(f'https://steamdb.info/app/{app}/info/')
    soup2 = bs(res.text, 'html.parser')
    try:
        if res.status_code != 200:
            print(f'banned - {idx, app}')
            raise Exception
    except:
        time.sleep(60)
        res = s.get(f'https://steamdb.info/app/{app}/info/')
        soup2 = bs(res.text, 'html.parser')
        
    x = {}
    for tr in soup2.select_one('div[id="info"]').select_one('tbody').select('tr'):
        a = tr.select('td')[0].text
        b = tr.select('td')[1].text
        x[a] = b
    
    for z_ in z:
        if z_ in x.keys():
            y[z_].append(x[z_])
        else:
            y[z_].append(np.nan)
```
