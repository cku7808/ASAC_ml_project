```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(m3[m3.columns[1:-1]], m3['Review'], test_size=0.3, random_state=1234, stratify=m3['Review'])
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1234)
```

```python
rf.fit(X_train, Y_train)
```

```python
pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(Y_test, pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold

str_kfold = StratifiedKFold(n_splits=10, random_state=1234, shuffle=True )

rf = RandomForestClassifier(random_state=1234)

scores = cross_val_score( rf, 
                        X_train, Y_train,
                        cv = str_kfold,
                        scoring = "accuracy")

for iter_count, acc in enumerate( scores):
    print(f"rf {iter_count}번째 검증 ACC:{acc}")
print("\nrf 모델 Accuracy 평균:", scores.mean())
print("rf 모델 Accuracy std:", scores.std())
```

```python
str_kfold = StratifiedKFold(n_splits=10, random_state=1234, shuffle=True )
```

```python
rf = RandomForestClassifier(random_state=1234)

n_iter = 10

parameters = {
    "n_estimators" : range(100, 501, 50),
    
}

rf_kfold_rgs = RandomizedSearchCV(rf,
                                param_distributions=parameters,
                                cv = str_kfold,
                                scoring='accuracy',
                                n_jobs=-1,
                                random_state=1234,
                                n_iter=n_iter)

rf_kfold_rgs.fit(X_train, Y_train)
```

```python
print(rf_kfold_rgs.best_score_)
```

```python
rf_kfold_rgs_best = rf_kfold_rgs.best_estimator_
rf_kfold_rgs_ypred = rf_kfold_rgs_best.predict(X_test)
rf_kfold_rgs_acc = accuracy_score(Y_test, rf_kfold_rgs_ypred)

print("rf kfold RGS ACC:",rf_kfold_rgs_acc)
```
