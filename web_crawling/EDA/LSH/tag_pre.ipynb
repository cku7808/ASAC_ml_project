```python
pwd = os.getcwd()
csv_list = glob(f'{pwd}\\**\\appid_*.csv', recursive=True)
print(csv_list)
```

```python
tag_set = set()

for csv in csv_list:
    raw = pd.read_csv(csv, usecols=['appid','tags'], sep='\t')
    
    raw['keys'] = raw['tags'].apply(lambda x: set(eval(x)) if x != '[]' else set())
    tag_set |= reduce(lambda x, y : x | y, raw['keys'])
            
```

```python
len(tag_set)
```

```python
from collections import deque
import numpy as np

for csv in csv_list:
    dir_name = os.path.split(csv)[0]
    base_name = os.path.splitext(os.path.split(csv)[1])[0]
    
    raw = pd.read_csv(csv, usecols=['appid','tags'], sep='\t')
    raw['keys'] = raw['tags'].apply(lambda x: set(eval(x)) if x != '[]' else set())
    raw['tags'] = raw['tags'].apply(lambda x: eval(x) if x != '[]' else set())
    
    tag_dict = {tag : deque([np.nan for i in range(len(raw))]) for tag in tag_set}
    
    for idx in range(len(raw)):
        print(f'{base_name} - {idx}')
        
        for key in raw.loc[idx, 'keys']:
            tag_dict[key][idx] = raw.loc[idx, 'tags'][key]
    
    tag_dict['appid'] = raw['appid']
    tag_df = pd.DataFrame(tag_dict)
    
    tag_df = tag_df[tag_df.columns[:-1].insert(0,tag_df.columns[-1])]
    tag_df.fillna(0, inplace=True)
    
    tag_df.to_csv(f'{dir_name}\\{base_name}_tag.csv', index=False)
    
```
