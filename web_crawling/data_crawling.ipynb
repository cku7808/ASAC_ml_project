```python
from collections import deque
import numpy as np
import time
```

```python

response = requests.get('https://steamdb.info/stats/gameratings/2023/', cookies=cookies, headers=headers)
soup = BeautifulSoup(response.text,"html.parser")

# datalist = soup.find("table",{"id":"table-apps"}).find("tbody").find_all("tr")
datalist = soup.find("table",{"id":"table-apps"}).find("tbody").find_all("tr")

Z = ["Developer","ValidOSList","community_visible_stats","Publisher",
            "IsFreeApp","Exclude from family sharing","has_adult_content",
            "Store Release Date","Primary Genre","Store Genres","Is free on the store?",
            "market_presence","review_score","review_percentage","SteamChinaApproved",
            "has_adult_content_violence","ReleaseStateSteamChina"]

Y = {}
for elem in Z:
    Y[elem] = deque([])

for idx,data in enumerate(datalist):
    if idx > 50:
        break
    appID = data["data-appid"]
    print(appID)

    base_url = "https://steamdb.info/app/"
    
    # chart
    # response = requests.get(base_url+appID+"/charts/", cookies=cookies, headers=headers)
    # soup = BeautifulSoup(response.text,"html.parser")
    # main = soup.find("ul",class_="app-chart-numbers-big").find_all("li")
    
    # playersRightNow = main[1].find("strong").text
    # twentyfourPeak = main[2].find("strong").text
    # allTimePeak = main[3].find("strong").text
    
    # app-info
    response = requests.get(base_url+appID+"/info/", cookies=cookies, headers=headers)
    soup = BeautifulSoup(response.text,"html.parser")
    main = soup.find("div",{"id":"info"}).find("table").find("tbody").find_all("tr")
    
    X = {}
    for col in main:
        tmp = col.find_all("td")
        X[tmp[0].text] = tmp[1].text
    # print(X)

    for elem in Z:
        if elem in X.keys():
            Y[elem].append(X[elem])
        else:
            Y[elem].append(np.nan)

    time.sleep(60)
print(Y)
```
