```python
from bs4 import BeautifulSoup
import requests
from collections import deque
import time
import numpy as np
import urllib.request
import json
```

```python
file_path = r"C:\Users\NT550-052\OneDrive\바탕 화면\ASAC\프로젝트\ML 프로젝트\appid_2022.txt"
file = open(file_path,"r")
```

```python
appids = []
while True:
    line = file.readline()
    if not line: break
    appids.append(int(line.strip()))
appids
```

```python
Z = ["required_age","is_free","controller_support","dlc","platforms","recommendations","achievements","release_date"]
```

```python
no_data = []
final_data = {}
```

```python
for idx,id in enumerate(appids):
    path = f"https://store.steampowered.com/api/appdetails?appids={id}"

    g_info = {}
    
    res = requests.get(path)
    # if res.status_code == 429: # 여기를 429두고 기다려, != > ==
    #     time.sleep(240)
    print(idx,id)
    # res = requests.get(path)
    try:
        data = res.json()
        # print(data[str(id)]["data"])
        
        if data[str(id)]["success"] == "true":
            keys = data[str(id)]["data"].keys()
            # print(data[str(id)]["data"].keys())
            main = data[str(id)]["data"]
            
            for elem in Z:
                if elem in keys:
                    if elem in ["recommendations","achievements"]:
                        g_info[elem] = main[elem]["total"]
                    elif elem == "release_date":
                        g_info[elem] = main[elem]["date"]
                    elif elem == "dlc":
                        g_info[elem] = len(main[elem])
                    else:
                        g_info[elem] = main[elem]
                else:
                    g_info[elem] = np.nan

            final_data[id] = g_info
        time.sleep(1)
    except:
        print(id,"error")
print(final_data)
    
```
