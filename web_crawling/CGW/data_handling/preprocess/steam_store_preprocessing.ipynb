```python
steam_df = pd.DataFrame(columns=["appid","required_age","is_free","controller_support","dlc","metacritic","windows","mac","linux","platforms_num","recommendations","achievements","release_date"])
steam_df
```

```python
def dateTransform(date):
    if type(date) == float:
        return np.nan
    
    date = re.sub(r"[,./ ]","",date)
    day = date[:2]
    year = date[-4:]
    month = date.replace(day,"").replace(year,"")
    tmp = [year,month,day]
    date = "-".join(tmp)

    return date
```

```python
for idx,item in enumerate(data.items()):
    key = item[0]
    value = item[1]
    row = [key]

    for k, v in value.items():
        if k == "platforms":
            if type(v) == dict:
                for pk in v.keys():
                    row.append(int(v[pk]))
                row.append(sum(v.values()))
            else:
                for _ in range(4):
                    row.append(np.nan)
                    
        elif k == "release_date":
            v = dateTransform(v)
            row.append(v)
        else:
            row.append(v)
    
    steam_df.loc[idx] = row
steam_df
```
