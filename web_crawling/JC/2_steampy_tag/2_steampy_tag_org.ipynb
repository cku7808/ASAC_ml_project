```python
raw['tags'] = raw['tags'].apply(lambda x : eval(x) if x != [] else set() ) # 셀이 빈 리스트가 아니면 eval(x), 빈리스트이면 set()?
raw['tags']
```

```python
# tags unqiue 값 정리
tag_set = set()
for i in range(len(raw)):
    tag_set |= set(raw.loc[i, 'tags']) 
    # |= 비트 연산자 : a와 b의 비트를 OR 연산한 후 결과를 a에 할당
    
```

```python
len(tag_set)
```

```python
from collections import deque
tag_dict = {tag : deque([]) for tag in tag_set}
```

```python
tag_dict
```

```python
# raw df 행을 순환하며 tag_set 을 순회하며, tag_dict에 각 tag 에 해당하는 셀의 값을 저장 ->  tag_dict는 각 태그에 대한 값들의 리스트
for i in range(len(raw)):
    for tag in tag_set:
        try:
            tag_dict[tag].append(raw.loc[i, 'tags'][tag])
        except:
            tag_dict[tag].append(0)
```

```python
tag_dict['appid'] = raw['appid']
```

```python
tag_df = pd.DataFrame(tag_dict)
```

```python
tag_df.describe()
```

```python
# column 순서 정리
col = ['appid'] + tag_df.columns[:-1].to_list()
tag_df = tag_df[col]
```
